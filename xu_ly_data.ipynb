```python
# Loại bỏ cột id
data = data.drop('id', axis=1)

# Kiểm tra dữ liệu sau khi loại bỏ cột id
print("Shape của dữ liệu sau khi loại bỏ cột id:", data.shape)
print("\nThông tin về dữ liệu:")
data.info()

# Kiểm tra giá trị null
print("\nKiểm tra giá trị null:")
print(data.isnull().sum())

# Hiển thị 5 dòng đầu tiên của dữ liệu đã được làm sạch
print("\n5 dòng đầu tiên của dữ liệu đã được làm sạch:")
data.head()
```

```python
# Loại bỏ cột 'id' và 'income'
data_moi = data.drop([ 'income'], axis=1)

# Kiểm tra kích thước của dữ liệu sau khi loại bỏ cột
print("Shape của dữ liệu sau khi loại bỏ cột id và income:", data_moi.shape)

# Thông tin tổng quan
print("\nThông tin về dữ liệu:")
data_moi.info()

# Kiểm tra giá trị null
print("\nKiểm tra giá trị null:")
print(data_moi.isnull().sum())

# Hiển thị 5 dòng đầu tiên
print("\n5 dòng đầu tiên của dữ liệu đã được làm sạch:")
data_moi.head()
```
